```python
from src.video_composition.static_background_image_component import StaticBackgroundImageComponent
from src.video_composition.composer import VideoComposition
from src.video_composition.text_component import TextComponent
from src.video_composition.timed_text_component import TimedTextComponent

background_start_time = 2  # Start at 2 seconds into the video
background_duration = 10  # Last for 10 seconds
text_start_time = 4  # Start at 4 seconds into the video
text_duration = 5  # Last for 5 seconds

width = 1024
height = 1024
fps = 30
font = 50
output_file = 'composition.mp4'
image = 'image.png'
```

```python
video_composition = VideoComposition(width=width, height=height, fps=fps)
video_composition.add_component(StaticBackgroundImageComponent(image, width=width, height=height, fps=fps, start_time=background_start_time, duration=background_duration))
video_composition.add_component(TextComponent(text="Hello", font=font, width=width, height=height, fps=fps, start_time=text_start_time, duration=text_duration))
video_composition.create_video(output_file=output_file)
```
