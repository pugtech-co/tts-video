```python
import json

# Assuming 'story' contains the JSON data as a string
# If you already have it as a Python dictionary, you can skip the json.loads step.
story_data = json.loads(story)

# Initialize an empty string to hold the concatenated NarratorText
concatenated_text = ""

for segment in story_data['EphemeraHollerTikTokStory']:
    narrator_text = segment['NarratorText']
    concatenated_text += narrator_text + " "

text = concatenated_text.rstrip()

# Now, 'concatenated_text' contains all the NarratorText as a single string.
print("Concatenated Narrator Text:")
print(concatenated_text)
```

```python
background_image_path = './images/barn.png'
audio_file_path = './music/creepy-trimmed.wav'


from src.video_composition.composer import VideoComposition

from src.video_composition.combo.tts_component import TTSComponent
from src.video_composition.audio.audio_file_component import AudioFileComponent
from src.video_composition.video.static_background_image_component import StaticBackgroundImageComponent

output_file = "output/horror.mp4"

tts = TTSComponent(text, color=(200, 10, 0))

video_composition = VideoComposition(width=1080, height=1920)
video_composition.add_component(StaticBackgroundImageComponent(background_image_path, width=1080, height=1920, fps=30,start_time=0, duration=tts.total_duration()))
video_composition.add_component(tts)
video_composition.add_component(AudioFileComponent(audio_file_path, start_time=0))

video_composition.create_video(output_file)

```
